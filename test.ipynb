```python
import numpy as np

def softmax_with_temperature(logits, T=1.0):
    """
    Compute the softmax with temperature.
    
    Parameters:
        logits: list or np.array of input scores (z values)
        T: temperature (float), T > 0
    
    Returns:
        Softmax probabilities (numpy array)
    """
    logits = np.array(logits)
    scaled_logits = logits / T
    exps = np.exp(scaled_logits - np.max(scaled_logits))  # stability trick
    return exps / np.sum(exps)

# Example usage
logits = [0.01, 0.03, 0.25, 0.01, 0.4, 0.1, 0.05, 0.15]
temperature = 1.0

softmax_scores = softmax_with_temperature(logits, T=temperature)
print(np.round(softmax_scores, 3))
```

```python
import numpy as np

def softmax_with_temperature(logits, T=1.0):
    logits = np.array(logits)
    scaled_logits = logits / T  # don't overwrite original
    exps = np.exp(scaled_logits - np.max(scaled_logits))  # stability
    return exps / np.sum(exps)

# Original logits
logits = [0.01, 0.03, 0.25, 0.01, 0.4, 0.1, 0.05, 0.15]

# Softmax at T = 1 (normal)
print("T=1:", np.round(softmax_with_temperature(logits, T=1.0), 3))

# Softmax at T = 2
print("T=2:", np.round(softmax_with_temperature(logits, T=2.0), 3))
```
